This is Air Passengers data from https://www.kaggle.com/rakannimer/air-passengers/. It has two column: Date and number of passengers in that month.  The data is from 1949 January to 1960 December

Importing All the libraries required

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

Reading the data as pandas dataframe into variable:df and then exporting it as numpy 1 Dimensional Array in variable: data

```python
df = pd.read_csv('./data.csv')
data = df['passengers'].values
```

Plotting the data and Displaying number of Rows and Column in data

```python
print(data.shape)
n = data.size
plt.plot(data)
plt.show()
```

Fit the data in Model Ridge():Model and perform curve fitting of degree = variable degree
Here, the curve shows the trend of the time series


```python
degree = 3

Model= make_pipeline(PolynomialFeatures(degree), Ridge())
Model.fit(np.array(range(1,n+1)).reshape(-1,1), data)

predicted_data = Model.predict(np.array(range(1,n+1)).reshape(-1,1))
plt.plot(np.array(range(1,n+1)).reshape(-1,1),data)
plt.plot(predicted_data)
plt.show()
```

```python
plt.plot(predicted_data)
plt.show()
```

Removing Trend from the data and putting it in detrended_data

```python
detrended_data = data- predicted_data
plt.plot(detrended_data)
plt.show()
```

```python
pp = n//50
normalized_detrended_data = np.zeros(pp*50,)
for k in range(pp):
    temp2 = np.std(detrended_data[k*50:(k+1)*50] )
    mean=np.mean(detrended_data[k*50:(k+1)*50])
    for g in range(50):
            normalized_detrended_data[k*50+g] = (detrended_data[k*50+g]-mean) /temp2

plt.plot(normalized_detrended_data)
plt.show()
```

```python
yf = np.fft.fft(normalized_detrended_data)
yf[0]=0
real_yf = 2.0/n * np.abs(yf[:n//2])
plt.plot(real_yf)
plt.show()

frequency = np.argmax(real_yf)+1
print(frequency)
```

```python
p = np.argpartition(real_yf, -3)[-3:]
q=real_yf[p]

def season(t):
    t = t%frequency
    temp1 =0
    for i in range(p.size):
        temp1 = temp1 + q[i]*np.sin((2*3.1415/p[i])*t)
    return temp1
    
season_data = np.zeros((n,1))
for m in range(n):
    season_data[m]=season(m)

plt.plot(season_data)
plt.show()


    
```

```python
f,(ax1, ax2,ax3)=plt.subplots(3,1)
ax1.plot(data)
ax2.plot(predicted_data)
ax3.plot(season_data)
plt.show()
```
